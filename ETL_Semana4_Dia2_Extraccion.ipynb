```python
import pandas as pd
import sqlite3
import json
```

```python
df_csv = pd.read_csv("ventas.csv")

print("Desde CSV:")
df_csv
```

```python
df_excel_ventas = pd.read_excel("datos.xlsx", sheet_name="Ventas")
df_excel_clientes = pd.read_excel("datos.xlsx", sheet_name="Clientes")

print("Desde Excel - Ventas:")
df_excel_ventas
```

```python
df_json = pd.read_json("productos.json")

print("Desde JSON:")
df_json
```

```python
conn = sqlite3.connect("ventas.db")

df_sql = pd.read_sql("SELECT * FROM pedidos", conn)

conn.close()

print("Desde SQLite:")
df_sql
```

```python
api_response = {
    "status": "success",
    "data": [
        {"id": 201, "producto": "Webcam", "stock": 15},
        {"id": 202, "producto": "Micrófono", "stock": 8}
    ]
}

df_api = pd.DataFrame(api_response["data"])

print("Desde API simulada:")
df_api
```

```python
dataframes = {
    "CSV": df_csv,
    "Excel Ventas": df_excel_ventas,
    "Excel Clientes": df_excel_clientes,
    "JSON": df_json,
    "SQLite": df_sql,
    "API": df_api
}

for fuente, df in dataframes.items():
    print(f"{fuente}: {df.shape}")
```

```python
with pd.ExcelWriter("evidencia_extraccion_dia2.xlsx", engine="openpyxl") as writer:
    df_csv.to_excel(writer, sheet_name="CSV_Ventas", index=False)
    df_excel_ventas.to_excel(writer, sheet_name="Excel_Ventas", index=False)
    df_excel_clientes.to_excel(writer, sheet_name="Excel_Clientes", index=False)
    df_json.to_excel(writer, sheet_name="JSON_Productos", index=False)
    df_sql.to_excel(writer, sheet_name="SQLite_Pedidos", index=False)
    df_api.to_excel(writer, sheet_name="API_Stock", index=False)
```
